# Visual Data Exploration with Matplotlib

--------------

```python
import numpy as np

import pandas as pd

%matplotlib inline

import matplotlib.pyplot as plt
```

```python
data1 = np.random.normal(0, 0.1, 1000)

data2 = np.random.normal(1, 0.4, 1000) + np.linspace(0, 1, 1000)

data3 = 2 + np.random.random(1000) * np.linspace(1, 5, 1000)

data4 = np.random.normal(3, 0.2, 1000) + 0.3 * np.sin(np.linspace(0, 20, 1000))
```

```python
data = np.vstack([data1, data2, data3, data4]).transpose()

data[:10]
```

```python
df = pd.DataFrame(data, columns = ['data1', 'data2', 'data3', 'data4'])

df.head(10)
```

-------

# 1. Line Plot

```python
df.plot(title = 'Line plot', figsize = (15, 7))
```

---------

# 2. Scatter Plot

```python
df.plot(style = '.', figsize = (15, 7))
```

```python
_ = df.plot(kind = 'scatter', x = 'data1', y = 'data2',
           xlim = (-1.5, 1.5), ylim = (0, 3), figsize = (15, 7))
```

-------

# 3. Histograms

```python
df.plot(kind='hist', # histogram
        bins=50, # number of bins
        title='Histogram', # title of the plot
        alpha=0.6, # control transparency
        figsize = (15, 7))
```

-----

# 4. Cumulative Distribution

```python
df.plot(kind = 'hist', # histogram
        bins = 100,
        title = 'Cumulative distributions',
        normed = True,
        cumulative = True,
        alpha = 0.4,
        figsize = (15, 7))
```

--------

# 5. Box Plot

```python
df.plot(kind = 'box', title = 'Boxplot', figsize = (15, 7))
```

--------

# 6. Subplots

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 7))

df.plot(ax    = ax[0][0], # first row first column
        title = 'Line plot')

df.plot(ax    = ax[0][1], # first row second column
        style = 'o',
        title = 'Scatter plot')

df.plot(ax    = ax[1][0], # second row first column
        kind  = 'hist',
        bins  = 50,
        title = 'Histogram')

df.plot(ax    = ax[1][1], # second row second column
        kind  = 'box',
        title = 'Boxplot')

plt.tight_layout() # for nice to read
```

-------

# 7. Pie Charts

```python
gt01 = df['data1'] > 0.1

piecounts = gt01.value_counts()

piecounts
```

```python
piecounts.plot(kind       = 'pie',
               figsize    = (15, 7),
               explode    = [0, 0.15],
               labels     = ['<= 0.1', '> 0.1'],
               autopct    = '%1.1f%%', # percantages
               shadow     = True,
               startangle = 90, # start ange : 90 degrees
               fontsize   = 16)
```

```python
piecounts.plot(kind       = 'pie',
               figsize    = (15, 7),
               fontsize   = 16)
```

# 8. Hexbin plot

```python
np.random.normal((0, 0), 2, size = (1000, 2))
```

```python
data = np.vstack([np.random.normal((0, 0), 2, size = (1000, 2)),
                  np.random.normal((9, 9), 3, size = (2000, 2))])

df = pd.DataFrame(data, columns=['x', 'y'])

df.head()
```

```python
df.plot(kind = 'kde', figsize = (15, 7))
```

```python
df.plot(kind = 'hexbin', x = 'x', y = 'y', bins = 100, cmap = 'rainbow', figsize = (15, 7))
```
