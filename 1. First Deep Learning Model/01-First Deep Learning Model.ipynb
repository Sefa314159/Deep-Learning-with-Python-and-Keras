# First Deep Learning Model

```python
import numpy as np
```

```python
%matplotlib inline  # '%' <-- cell magic :  adds functionality to the jupyter
import matplotlib.pyplot as plt
```

```python
# then we import a generator of samples from sklearn

from sklearn.datasets.samples_generator import make_circles
```

```python
X, y = make_circles(n_samples = 1000,
                   noise = 0.1,
                   factor = 0.2,
                   random_state = 0)
```

```python
X.shape # 1000 samples, 2 columns
```

```python
y.shape
```

```python
X
```

```python
y[0:10]
```

```python
# this is our trainin data

plt.figure(figsize = (5, 5))
plt.plot(X[y == 0, 0], X[y == 0, 1], 'ob', alpha = 0.5)
plt.plot(X[y == 1, 0], X[y == 1, 1], 'xr', alpha = 0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses")
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```python
model = Sequential()

model.add(Dense(4, input_shape = (2, ), activation = 'tanh'))

model.add(Dense(1, activation = 'sigmoid'))

model.compile(SGD(lr = 0.5), 'binary_crossentropy', metrics = ['accuracy'])

model.fit(X, y, epochs = 20)
```

```python
hticks = np.linspace(-1.5, 1.5, 101)

vticks = np.linspace(-1.5, 1.5, 101)

aa, bb = np.meshgrid(hticks, vticks)

ab     = np.c_[aa.ravel(), bb.ravel()]

c      = model.predict(ab)

cc     = c.reshape(aa.shape)
```

```python
plt.figure(figsize = (5, 5))

plt.contourf(aa, bb, cc, cmap = 'bwr', alpha = 0.2)

plt.plot(X[y == 0, 0], X[y == 0, 1], 'ob', alpha = 0.2)

plt.plot(X[y == 1, 0], X[y == 1, 1], 'xr', alpha = 0.5)

plt.xlim(-1.5, 1.5)

plt.ylim(-1.5, 1.5)

plt.legend(['0', '1'])

plt.title("Blue circles and Red crosses")
```
